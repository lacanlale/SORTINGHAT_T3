```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data/model_predictions/rf_false.csv")
df = df.drop(['Index'], axis=1)
df.describe()
```

```python
df.head(5)
```

```python
def get_des(label, df, func):
    ex_df = df.loc[df['y_act'] == label]
    return func(ex_df)
```

```python
labels = df.y_act.unique()
funcs = [np.mean, np.std]
stats_dict = {}
for label in labels:
    stats_dict[label] = {}
    for func in funcs:
        stats_dict[label][func.__name__] = get_des(label, df, func)
stats_dict
```
