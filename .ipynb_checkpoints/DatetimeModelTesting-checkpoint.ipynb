# Rulebook for labelling
Rules/guidelines/cases for inferring data types that have been labelled “needs extraction”

1. **Numbers**

<u>Case a.</u> Number value proceeded and/or preceded by a unit string, as well as possibly in-between e.g. ‘100 inches, ‘$15’, ‘-0.12 amps’, ‘~23 feet’, ‘US$ PER 100 LBS’, ‘$5,000,000.00’

2. **List**

<u>Case b.</u> 
Text corpus that contains a delimiter that differs from standard date formats and/or may represent a variable type with the intention for computer processing e.g ‘a,b,c’, ‘word 1; word 2; word 3’, ‘{men|clothing, women|clothing, children|toys etc}’, ‘{'id': 7, 'name': 'Funny', 'count': 19645}’ 

3. **Datetime**

<u>Case c.</u> 
Follows standard numerical formatting seen in applications e.g. ‘12.22.16’, ‘1998/01/01’, ‘30-8-2002’

<u>Case d.</u> Uses a combination of strings and numbers to convey a date e.g. ‘December 1’, ‘October 30, 2019’, ‘The first of May’

<u>Case e.</u> Follows standard numerical formatting seen in applications e.g. ‘12:30, ‘1:30:20’, ‘15:00’

<u>Case f.</u> Uses a combination of strings and numbers to convey a time of day e.g. ‘Half past 12’, ‘12 PM’, ‘9 o’ clock in the morning’

4. **Sentence**

<u>Case g.</u> Text corpus that does not qualify as a date, timestamp, or URL and conveys significant contextual meaning based on the dataset (title, description, individual information, text excerpt, etc.). e.g. ‘Do schools kill creativity?’, ‘SEE TEXT’, ‘statistics, computer science’ This text should also not represent a variable type that may be intended for computer processing.

5. **URL**

<u>Case h.</u> Contains a protocol prefix and/or top-level domain suffix with possible additional categorical/product information e.g. ‘google.com’, ‘https://www.ucsd.edu/’, ‘https://www.researchgate.net/blog’

6. **Custom Object**

<u>Case i.</u> Multiple numerical values whose context is unrelated to date or time and may require more contextual processing than other numbers e.g. \'(-0.022, 2.2]\', \'(-5.0000, 33.0000)\'

<u>Case j.</u> Phrases containing words that convey specific geographic information e.g. ‘address : 9415 Campus Point Dr’, ‘mailing address: 345 Airport Rd PO Box 1242 Malta’, ‘MT 59538’

<u>Case k.</u> Electronic identification data e.g. , ‘email : johndoe@website.com’

<u>Case l.</u> Descriptive data that should have been classified by an earlier step.

7. **Unusable**

<u>Case m.</u> Total of NaN values is too high and/or sample items do not convey contextual background of what the data is

## Notes about categories
- **Datetime**
    - Before being called *datetime*, *date* and *time* were separate categories. They were later combined because of (1) low sample numbers and (2) large amounts of entries had both date and time as a single value. 
    - Some columns will have different formats of a value
        - *Actual example from <u>Time of Issue from CGB_-_Consumer_Complaints_Data.csv</u>*
            - "9:35 a.m." and "1:00 pm"
    - Columns differ in how their dates are formatted, using either '/', '-', or none.
- **Numbers**
    - Some columns have values that distinguish them from being numerical values
        - *Actual example from <u>duration from pitStops.csv</u>*
            - Majority look similar to '26.898' but also has values similar to '1:08.357'
- **List**
    - Some lists look similar to programming arrays or dicts
        - *Actual Example from <u>Reviews @ TA_restaurants_curated.csv</u>*
            - [['Excellent Breakfast'], ['08/22/2012']]
    - Delimiters vary (seen: *','*, *';'*, *'>'*, ' ')

# Model Testing with numerical features 

````python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
from sklearn import neighbors
from sklearn import svm
from sklearn import linear_model
from sklearn import tree
from sklearn import naive_bayes
from sklearn import ensemble
from sklearn.metrics import confusion_matrix
````

````python
def get_param_report(X, y, est, params, name):
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.5, random_state=0)
    print(f"> Tuning hyper-parameters for {name}")
    print()

    clf = GridSearchCV(est, params, cv=5)
    clf.fit(X_train, y_train)

    print("> Best parameters set found on development set:")
    print()
    print(clf.best_params_)
    print()
    print("> Grid scores on development set:")
    print()
    means = clf.cv_results_['mean_test_score']
    stds = clf.cv_results_['std_test_score']
    for mean, std, params in zip(means, stds, clf.cv_results_['params']):
        print("> %0.3f (+/-%0.03f) for %r"
              % (mean, std * 2, params))
    print()

    print("> Detailed classification report:")
    print()
    print("> The model is trained on the full development set.")
    print("> The scores are computed on the full evaluation set.")
    print()
    y_true, y_pred = y_test, clf.predict(X_test)
    print(classification_report(y_true, y_pred))
    print()
    

def get_pred_perc(model_list, folder):
    path = f'data/needs_extraction_data/classified_csvs/datetime/{folder}'
    acc = {}
    mapping = {
        "['Datetime']" :'Datetime',
        "['Sentence']" :'Sentence',
        "['Custom Object']" :'Custom Object',
        "['Unusable']" :'Unusable',
        "['URL']" :'URL',
        "['Numbers']" :'Numbers',
        "['List']" :'List'
    }
    for model in model_list:
        df = pd.read_csv(f'{path}/{model}.csv')
        df.replace({'y_pred' : mapping}, inplace=True)
        df.to_csv(f'{path}/{model}.csv', index=False)
        correct = len(df.loc[df.y_act == df.y_pred])
        acc[model] = (correct/len(df))*100
    return acc


def make_labelled_csv(clf, model_name, x, y):
    path = 'data/needs_extraction_data/classified_csvs/datetime/numerical_features_only/'
    x_train, x_test, y_train, y_test = train_test_split(
            x, y, test_size=0.1)
    clf.fit(x_train, y_train)
    df = pd.read_csv('data/needs_extraction_data/date_and_time_combined.csv')
    preds = []
    cut = df[['mean_word_count','std_dev_word_count','has_delimeters']]
    for row in cut.itertuples():
        preds.append(clf.predict(np.array([row[1], row[2], row[3]]).reshape(1, -1)))
    df.y_pred = preds
    df.to_csv(f'{path}{model_name}.csv', index=False)
````

````python
encoder = {0: 'Datetime', 1: 'Sentence', 2: 'Custom Object', 3: 'Unusable', 4: 'URL', 5: 'Numbers', 6: 'List'}
````

````python
from yellowbrick.classifier import ConfusionMatrix
from yellowbrick.classifier import ClassBalance
from yellowbrick.classifier import ClassificationReport


def test_mult(x, y, clf, num=100):
    '''
    Test a model #NUM times

    :x: x-values
    :y: y-values
    :clf: classifier
    :num: # of times to be tested
    :returns: Average accuracy (rounded to 2 decimal places)
    '''
    acc_attempts = []
    x_train, x_test, y_train, y_test = 0, 0, 0, 0
    for _ in range(num):
        x_train, x_test, y_train, y_test = train_test_split(
            x, y)
        clf.fit(x_train, y_train)
        acc_attempts.append(clf.score(x_test, y_test)*100)

    cm = ConfusionMatrix(clf,
                         percent=True,
                         classes=clf.classes_,
                         label_encoder=encoder)
    cm.fit(x_train, y_train)
    cm.score(x_test, y_test)
    cm.poof()

    balance = ClassBalance(clf, 
                           percent=True,
                           classes=clf.classes_,
                           label_encoder=encoder)
    balance.fit(x_train, y_train)
    balance.score(x_test, y_test)
    balance.poof()

    report = ClassificationReport(clf, 
                                  percent=True,
                                  classes=clf.classes_,
                                  label_encoder=encoder)
    report.fit(x_train, y_train)
    report.score(x_test, y_test)
    report.poof()

    return np.mean(acc_attempts)
````

````python
# df.loc[df.y_act == 'Datetime', 'y_act'] = 0
# df.loc[df.y_act == 'Sentence', 'y_act'] = 1
# df.loc[df.y_act == 'Custom Object', 'y_act'] = 2
# df.loc[df.y_act == 'Unusable', 'y_act'] = 3
# df.loc[df.y_act == 'URL', 'y_act'] = 4
# df.loc[df.y_act == 'Numbers', 'y_act'] = 5
# df.loc[df.y_act == 'List', 'y_act'] = 6

x = np.array(df.drop(['Record_id', 'Attribute_name', 'y_pred', 'y_act', 
                     'Reason', 'total_vals', 'num_nans', '%_nans',
                     'sample_1', 'sample_2', 'sample_3', 'sample_4', 'sample_5'], axis=1))
y = df['y_act']
````

## Finding best hyper-parameters

````python
sep = "================================================================"
# KNN
clf = neighbors.KNeighborsClassifier()
params = [{'n_neighbors': [5, 50, 100, 150], 'algorithm': [
    'ball_tree', 'kd_tree'], 'leaf_size': [10, 20, 30, 50, 100]}]
get_param_report(x, y, clf, params, 'KNN')
print(sep)

# SVM
clf = svm.SVC()
params = [{'kernel': ['rbf'], 'gamma': [1e-3, 1e-4],
           'C': [1, 10, 100]}, {'kernel': ['linear'], 'C': [1, 10, 100]}]
get_param_report(x, y, clf, params, 'SVM')
print(sep)

# Decision Tree
clf = tree.DecisionTreeClassifier()
params = {'splitter': ['best', 'random'], 'max_depth': [
    30, 50, 100, 500], 'min_samples_split': [2, 10, 50, 100]}
get_param_report(x, y, clf, params, 'Decision Tree')
print(sep)

# Random Forest
clf = ensemble.RandomForestClassifier()
params = {'n_estimators': [5, 10, 50], 'criterion': [
    'gini', 'entropy'], 'max_depth': [30, 50, 100, 500, 1000]}
get_param_report(x, y, clf, params, 'Random Forest')
print(sep)

# Logistic Regression
clf = linear_model.LogisticRegression()
params = {'penalty': ['l1', 'l2'], 'C': [1.0, 1.5, 2.0]}
get_param_report(x, y, clf, params, 'Logistic Regression')
print(sep)
````

````python
accuracy_perc_num = {}

# KNN
clf = neighbors.KNeighborsClassifier(algorithm='ball_tree', leaf_size=10, n_neighbors=5)
accuracy_perc_num['knn'] = test_mult(x, y, clf, 100)
make_labelled_csv(clf, 'knn', x, y)

# SVM
clf = svm.SVC(C=100, kernel='rbf')
accuracy_perc_num['svm'] = test_mult(x, y, clf, 100)
make_labelled_csv(clf, 'svm', x, y)

# Decision Tree
clf = tree.DecisionTreeClassifier(max_depth=50, min_samples_split=2, splitter='best')
accuracy_perc_num['dt'] = test_mult(x, y, clf, 100)
make_labelled_csv(clf, 'decision_tree', x, y)

# Random Forest
clf = ensemble.RandomForestClassifier(criterion='gini', max_depth=50, n_estimators=6)
accuracy_perc_num['rf'] = test_mult(x, y, clf, 100)
make_labelled_csv(clf, 'random_forest', x, y)

# Logistic Regression
clf = linear_model.LogisticRegression(C=1.0, penalty='l1')
accuracy_perc_num['lr'] = test_mult(x, y, clf, 100)
make_labelled_csv(clf, 'logistic_regression', x, y)

# Naive Bayes (Gaussian)
clf = naive_bayes.GaussianNB()
accuracy_perc_num['nb_g'] = test_mult(x, y, clf, 100)
make_labelled_csv(clf, 'gauss_naive_bayes', x, y)

# Naive Bayes (Bernoulli)
clf = naive_bayes.BernoulliNB()
accuracy_perc_num['nb_b'] = test_mult(x, y, clf, 100)
make_labelled_csv(clf, 'bern_naive_bayes', x, y)

# Naive Bayes (Multi)
clf = naive_bayes.MultinomialNB()
accuracy_perc_num['nb_m'] = test_mult(x, y, clf, 100)
make_labelled_csv(clf, 'multi_naive_bayes', x, y)
````

## Comparing Accuracies

````python
def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width()/2., 1.05*height,
                '%d' % int(height),
                ha='center', va='bottom')
        
ind = np.arange(8)

plt.figure(figsize=(18, 10))

accbars = plt.bar(ind, [accuracy_perc_num[value] for value in accuracy_perc_num], color='g')
plt.axhline(0, color='black')
plt.margins(0.01)
plt.title('Accuracy Comparison')
plt.xticks(ind,('KNN', 'SVM', 'DECISION TREES', 'RANDOM FOREST', 'LOGISTIC REGRESSION', 'GAUSSIAN NAIVE BAYES', 'BERNOULLI NAIVE BAYES', 'MULTINOMIAL NAIVE BAYES'))
plt.yticks(np.linspace(0, 100, num=11))
plt.ylabel('Accuracy (%)')
plt.xlabel('Classifier')

autolabel(accbars)

plt.show()
````

## Additional feature combination testing

Features being used:
- mean_word_count
- std_word_count
- has_delimiters

````python
import itertools


def test_mult(x, y, clf, num=100):
    '''
    Test a model #NUM times

    :x: x-values
    :y: y-values
    :clf: classifier
    :num: # of times to be tested
    :returns: Average accuracy (rounded to 2 decimal places)
    '''
    acc_attempts = []
    x_train, x_test, y_train, y_test = 0, 0, 0, 0
    for _ in range(num):
        x_train, x_test, y_train, y_test = train_test_split(
            x, y, test_size=0.1)
        clf.fit(x_train, y_train)
        acc_attempts.append(clf.score(x_test, y_test)*100)
    return np.mean(acc_attempts)


def get_combs(iterable):
    combs = []
    for count in range(1, len(iterable)):
        for comb in itertools.combinations(iterable, count):
            combs.append(comb)
    return combs


def test_combs(combs, comb_names):
    accuracy = {}
    for name in comb_names:
        accuracy[name] = {}
        for comb in combs:
            x = np.array(df.drop(['Record_id', 'Attribute_name', 'y_pred', 'y_act',
                                  'Reason', 'total_vals', 'num_nans', '%_nans',
                                  'sample_1', 'sample_2', 'sample_3', 'sample_4', 'sample_5'] + list(comb), axis=1))
            # KNN
            clf = neighbors.KNeighborsClassifier(
                algorithm='ball_tree', leaf_size=30, n_neighbors=5)
            accuracy[name]['knn'] = test_mult(x, y, clf, 100)

            # SVM
            clf = svm.SVC(C=100, kernel='rbf')
            accuracy[name]['svm'] = test_mult(x, y, clf, 100)

            # Decision Tree
            clf = tree.DecisionTreeClassifier(
                max_depth=500, min_samples_split=2, splitter='best')
            accuracy[name]['dt'] = test_mult(x, y, clf, 100)

            # Random Forest
            clf = ensemble.RandomForestClassifier(
                criterion='gini', max_depth=30, n_estimators=10)
            accuracy[name]['rf'] = test_mult(x, y, clf, 100)

            # Logistic Regression
            clf = linear_model.LogisticRegression(C=1.0, penalty='l1')
            accuracy[name]['lr'] = test_mult(x, y, clf, 100)

            # Naive Bayes (Gaussian)
            clf = naive_bayes.GaussianNB()
            accuracy[name]['nb_g'] = test_mult(x, y, clf, 100)

            # Naive Bayes (Bernoulli)
            clf = naive_bayes.BernoulliNB()
            accuracy[name]['nb_b'] = test_mult(x, y, clf, 100)

            # Naive Bayes (Multi)
            clf = naive_bayes.MultinomialNB()
            accuracy[name]['nb_m'] = test_mult(x, y, clf, 100)
    return accuracy
````

````python
feats = ['mean_word_count', 'std_dev_word_count', 'has_delimeters']
comb_names = ['STDEV+HAS_DEL','MEAN+HAS_DEL','MEAN+STDEV','HAS_DEL','STDEV','MEAN']
combs = get_combs(feats)
accs = test_combs(combs, comb_names)
````

````python
def test_mult(x, y, clf, num=100):
    '''
    Test a model #NUM times

    :x: x-values
    :y: y-values
    :clf: classifier
    :num: # of times to be tested
    :returns: Average accuracy (rounded to 2 decimal places)
    '''
    acc_attempts = []
    x_train, x_test, y_train, y_test = 0,0,0,0
    for _ in range(num):
        x_train, x_test, y_train, y_test = train_test_split(
            x, y)
        Tfidf_vect = TfidfVectorizer()
        Tfidf_vect.fit(df['text_final'])
        
        Train_X_Tfidf = Tfidf_vect.transform(x_train)
        Test_X_Tfidf = Tfidf_vect.transform(x_test)
        clf.fit(Train_X_Tfidf, y_train)
        acc_attempts.append(clf.score(Test_X_Tfidf, y_test)*100)
    
    cm = ConfusionMatrix(
        clf,
        percent=True,
        classes=['Numbers', 'List', 'Datetime', 'Sentence', 'URL', 'Custom Object', 'Unusable'])
    cm.fit(Train_X_Tfidf, y_train)
    cm.score(Test_X_Tfidf, y_test)
    cm.poof()

    balance = ClassBalance(clf, classes=[
                           'Numbers', 'List', 'Datetime', 'Sentence', 'URL', 'Custom Object', 'Unusable'])
    balance.fit(Train_X_Tfidf, y_train)
    balance.score(Test_X_Tfidf, y_test)
    balance.poof()

    report = ClassificationReport(clf, classes=[
                                  'Numbers', 'List', 'Datetime', 'Sentence', 'URL', 'Custom Object', 'Unusable'])
    report.fit(Train_X_Tfidf, y_train)
    report.score(Test_X_Tfidf, y_test)
    report.poof()
    
    return np.mean(acc_attempts)


def get_param_report(X, y, est, params, name):
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.5, random_state=0)
    print(f"> Tuning hyper-parameters for {name}")
    print()
    
    Tfidf_vect = TfidfVectorizer()
    Tfidf_vect.fit(df['text_final'])
    Train_X_Tfidf = Tfidf_vect.transform(X_train)
    Test_X_Tfidf = Tfidf_vect.transform(X_test)
        
    clf = GridSearchCV(est, params, cv=5)
    clf.fit(Train_X_Tfidf, y_train)

    print("> Best parameters set found on development set:")
    print()
    print(clf.best_params_)
    print()
    print("> Grid scores on development set:")
    print()
    means = clf.cv_results_['mean_test_score']
    stds = clf.cv_results_['std_test_score']
    for mean, std, params in zip(means, stds, clf.cv_results_['params']):
        print("> %0.3f (+/-%0.03f) for %r"
              % (mean, std * 2, params))
    print()

    print("> Detailed classification report:")
    print()
    print("> The model is trained on the full development set.")
    print("> The scores are computed on the full evaluation set.")
    print()
    y_true, y_pred = y_test, clf.predict(Test_X_Tfidf)
    print(classification_report(y_true, y_pred))
    print()
    
    
def make_labelled_csv(clf, model_name, x, y, fit):
    path = 'data/needs_extraction_data/classified_csvs/datetime/attribute_name_only/'
    x_train, x_test, y_train, y_test = train_test_split(
            x, y, test_size=0.1)
    Tfidf_vect = TfidfVectorizer()
    Tfidf_vect.fit(fit)
    Train_X_Tfidf = Tfidf_vect.transform(x_train)
    clf.fit(Train_X_Tfidf, y_train)
    
    df = pd.read_csv('data/needs_extraction_data/date_and_time_combined.csv')
    cut = df[['Attribute_name']]
    preds = []
    for row in cut.itertuples():
        pred = clf.predict(Tfidf_vect.transform([row[1]]))
        preds.append(pred)
    df.y_pred = preds
    df.to_csv(f'{path}{model_name}.csv', index=False)
````

- Numbers : floating, mixed-integer, mixed-integer-float, decimal, complex
- List : string, mixed-integer, mixed-integer-float

````python
def get_infer_acc(model_list, folder):
    path = f'data/needs_extraction_data/classified_csvs/datetime'
    acc = {}
    for model in model_list:
        df = pd.read_csv(f'{path}/{folder}/{model}.csv')
        infers = df.pd_infer_type
        y_acts = list(df.y_act)
        correct = 0
        for row_index, val in infers.iteritems():
            vals = y_acts[row_index].split(' ')
            if val in vals:
                correct += 1
#                 print(f"[GOOD] pd.infer_dtype: {val} || row: {vals}")
#             else:
#                 print(f"[BAD] pd.infer_dtype: {val} || row: {vals}")
        acc[model] = (correct/len(df))*100
    return acc


def get_matching_acc(model_list, folder):
    path = f'data/needs_extraction_data/classified_csvs/datetime'
    acc = {}
    for model in model_list:
        df = pd.read_csv(f'{path}/{folder}/{model}.csv')
        infers = df.pd_infer_type
        y_preds = list(df.y_pred)
        correct = 0
        for row_index, val in infers.iteritems():
            vals = y_preds[row_index].split(' ')
            if val in vals:
                correct += 1
#                 print(f"[GOOD] pd.infer_dtype: {val} || row: {vals}")
#             else:
#                 print(f"[BAD] pd.infer_dtype: {val} || row: {vals}")
        acc[model] = (correct/len(df))*100
    return acc
````

````python
def map_cats(model_list, folder):
    path = f'data/needs_extraction_data/classified_csvs/datetime'
    mapping = {
        'Datetime' : "date datetime datetime64 time timedelta timedelta64 period",
        'Sentence' : "string mixed-integer mixed-integer-float",
        'Custom Object' : "string mixed categorical boolean unicode",
        'Unusable' : "string",
        'URL' : "string",
        'Numbers' : "floating mixed-integer mixed-integer-float decimal complex",
        'List' : "string mixed-integer mixed-integer-float"
    }
    for model in model_list:
        df = pd.read_csv(f'{path}/{folder}/{model}.csv')
        df.replace({'y_act' : mapping}, inplace=True)
        df.replace({'y_pred': mapping}, inplace=True)
        df.to_csv(f'{path}/{folder}/{model}.csv', index=False)
````

````python
def add_infer(model_list, folder):
    path = f'data/needs_extraction_data/classified_csvs/datetime'
    for model in model_list:
        infer = []
        df = pd.read_csv(f'{path}/{folder}/{model}.csv').fillna('NANVAL!')
        try:
            cut = df.drop(['Record_id', 'Attribute_name', 'y_pred', 'y_act', 'Reason', 'total_vals',
                       'num_nans', '%_nans', 'mean_word_count', 'std_dev_word_count', 'has_delimeters', 'pd_infer_type'], axis=1)
        except KeyError:
            cut = df.drop(['Record_id', 'Attribute_name', 'y_pred', 'y_act', 'Reason', 'total_vals',
                       'num_nans', '%_nans', 'mean_word_count', 'std_dev_word_count', 'has_delimeters'], axis=1)
        for samples in cut.itertuples():
            combined_samples = list(samples)[1:]
            infer.append(pd.api.types.infer_dtype(combined_samples))

        df['pd_infer_type'] = infer
        df.to_csv(f'{path}/{folder}/{model}.csv', index=False)
````

*Note*: Pandas infer_dtype only correctly identifies a variables type <u>**if that variable has been previously casted**</u>. Since infer_dtype is called on items read from a csv, all types are strings and are inferred as strings.
